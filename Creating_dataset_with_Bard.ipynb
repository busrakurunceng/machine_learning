```python
import re


# 'content' anahtarına sahip değeri al
content = weakness.get('content', '')

# Cümleleri ayırmak için uygun bir regex kullan
sentences = re.split(r'(?<=[.!?])\s+', content)

# Cümleleri tek tek yazdır veya kaydet
for i, sentence in enumerate(sentences):
    print(f'Cümle {i + 1}: {sentence.strip()}')
    # Burada cümleleri istediğiniz şekilde kaydedebilirsiniz.
```

```python
# 'content' anahtarından cümleleri al
metin = weakness['content']

# Başlık ifadelerini tanımla
fiziksel_halsizlik_basligi = "**Fiziksel halsizlik ifade eden cümleler**"
zihinsel_halsizlik_basligi = "**Zihinsel halsizlik ifade eden cümleler**"
diger_halsizlik_basligi = "**Diğer halsizlik ifade eden cümleler**"


# Fiziksel ve zihinsel halsizlik cümlelerini ayır
fiziksel_halsizlik_cümleleri = []
zihinsel_halsizlik_cümleleri = []
diger_halsizlik_cümleleri = []

lines = metin.split('\n')
current_list = None

for line in lines:
    if line.startswith(fiziksel_halsizlik_basligi):
        current_list = fiziksel_halsizlik_cümleleri
    elif line.startswith(zihinsel_halsizlik_basligi):
        current_list = zihinsel_halsizlik_cümleleri
    elif  line.startswith(diger_halsizlik_basligi):
        current_list = diger_halsizlik_cümleleri
    elif current_list is not None and line.strip():  # Boş satırları atla
        current_list.append(line)

# Sonuçları yazdır
print("Fiziksel Halsizlik Cümleleri:")
print("\n".join(fiziksel_halsizlik_cümleleri))

print("\nZihinsel Halsizlik Cümleleri:")
print("\n".join(zihinsel_halsizlik_cümleleri))

print("\nDiğer Halsizlik Cümleleri:")
print("\n".join(diger_halsizlik_cümleleri))
```
